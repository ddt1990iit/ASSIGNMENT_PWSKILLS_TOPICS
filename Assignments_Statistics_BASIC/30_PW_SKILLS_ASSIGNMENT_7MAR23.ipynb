## Q6. What is a Venn diagram?


## What is a Venn diagram?

**A Venn diagram** illustrates the relationships between two or more data sets. Venn diagrams are especially useful for highlighting similarities and differences and are commonly used to compare and contrast the characteristics of different data sets.

In a **Venn diagram**, circles are used to represent each data set. Inside each circle is a list of characteristics that define that data. Characteristics shared between two or more data sets are listed in the area where the circles overlap. Simple Venn diagrams consist of two overlapping circles, but complex Venn diagrams may compare up to five or more data sets using up to five or more circles.

**Venn diagrams** are commonly used in business and education settings to visualize and explore how things relate. In a business context, Venn diagrams are commonly used in product development, marketing, management, and more.




## Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:
### (i) A  B
### (ii) A  B


```python
# Solving the above problem 

# Assigning the above mentioned sets in two variable

set_A = {2,3,4,5,6,7}
set_B = {0,2,6,8,10}

print('set_A',set_A)
print('set_B',set_B)
print('\n')

# (i) A  B

intersection = set_A.intersection(set_B)

print('A  B = ',intersection)

# (i) A U B

union = set_A.union(set_B)

print('\nA  B = ',union)
```

## Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

```python
# Example code to explain how the outlier effect the central tendency and dispersion

# Consider the dataset of the height mentioned in above question

import numpy as np
import statistics

height_dataset = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

def central_tendency(list):
    mean = np.mean(list)
    median = np.median(list)
    mode = statistics.mode(list)
    std = statistics.stdev(list)
    return mean,median,mode,std

mean_simple,median_simple,mode_simple,std_simple = central_tendency(height_dataset)

print('The mean of the dataset without outlier',mean_simple)
print('The median of the dataset without outlier',median_simple)
print('The mode of the dataset without outlier',mode_simple)
print('The standard deviation of the dataset without outlier',std_simple)


    
        
```

```python
# Consider the dataset of the height with an outlier of height 1000,2000,550

height_outlier = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5,1000,2000,550]

mean_outlier,median_outlier,mode_outlier,std_outlier = central_tendency(height_outlier)

print('The mean of the dataset with outlier',mean_outlier)
print('The median of the dataset with outlier',median_outlier)
print('The mode of the dataset with outlier',mode_outlier)
print('The standard deviation of the dataset with outlier',std_outlier)
```

```python
# Adding both the mean into the dataframe to compare the value

import pandas as pd

index = ['mean','median','mode','standard deviation']

df =pd.DataFrame({'central_tendency_without_oulier':[mean_simple,median_simple,mode_simple,std_simple],
                 'central_tendency_with_outlier':[mean_outlier,median_outlier,mode_outlier,std_outlier]},index=index)

```

```python
display(df)
```

```python
# lets compare both of the central tendency with outlier and without outlier

df['difference'] = df['central_tendency_with_outlier']-df['central_tendency_without_oulier']

display(df)
```
